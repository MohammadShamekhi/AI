```python
import numpy as np
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# f(x) = x^2
# X_train = np.sort(np.random.uniform(0, 10, (100,1)), axis=0)
# y_train = X_train * X_train
# X_test = np.sort(np.random.uniform(0, 10, (100,1)), axis=0)
# y_test = X_test * X_test

# f(x) = sin(2x + 5)
X_train = np.sort(np.random.uniform(0, 10, (600,1)), axis=0)
y_train = np.sin(2 * X_train + 5)
X_test = np.sort(np.random.uniform(0, 10, (100,1)), axis=0)
y_test = np.sin(2 * X_test + 5)

# f(x) = e^(x)
# X_train = np.sort(np.random.uniform(-10, 10, (100,1)), axis=0)
# y_train = np.exp(X_train)
# X_test = np.sort(np.random.uniform(-10, 10, (100,1)), axis=0)
# y_test = np.exp(X_test)

# add noise to y_train
noise = np.random.normal(0, 1, size=y_train.shape)
y_train = y_train + 3 * noise

# Create an MLPClassifier model
mlp = MLPRegressor(hidden_layer_sizes=(300, 300, 300, 300, 300), max_iter=1000000)

# Train the model on the training data
mlp.fit(X_train, y_train)

# Make predictions on the test data
y_pred = mlp.predict(X_test)

# Calculate the MSE of the model
mse = mean_squared_error(y_test, y_pred)
print("MSE: ", mse)

# Create a scatter plot of the training data
plt.scatter(X_train, y_train)

# Create a scatter plot of the test data
plt.scatter(X_test, y_test)

# Create a scatter plot of the predicted data
plt.scatter(X_test, y_pred)

# add the legend and labels
plt.legend(["Training data", "Test data", "Predicted data"])
plt.xlabel("X")
plt.ylabel("Y")

#show the plot
plt.show()
```
