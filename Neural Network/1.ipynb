```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load data
data=pd.read_csv('Airplane.csv')

# Mapping Gender
data["Gender"] = data["Gender"].map( {'Female': 1, 'Male': 0} ).astype(int)

# Mapping Customer Type
data["Customer Type"] = data["Customer Type"].map( {'Loyal Customer': 1, 'disloyal Customer': 0} ).astype(int)

# Mapping Age
data.loc[ data['Age'] <= 26.5, 'Age'] 					     = 0
data.loc[(data['Age'] > 26.5) & (data['Age'] <= 46.0), 'Age'] = 1
data.loc[(data['Age'] > 46.0) & (data['Age'] <= 65.5), 'Age'] = 2
data.loc[data['Age'] > 65.5, 'Age']                            = 3

# Mapping Type of Travel
data["Type of Travel"] = data["Type of Travel"].map( {'Business travel': 1, 'Personal Travel': 0} ).astype(int)

# Mapping Type of Class
data["Class"] = data["Class"].map({'Business': 3, 'Eco': 1, 'Eco Plus': 0})

# Mapping Flight Distance
data.loc[ data['Flight Distance'] <= 1269.0, 'Flight Distance'] 					                     = 0
data.loc[(data['Flight Distance'] > 1269.0) & (data['Flight Distance'] <= 2507.0), 'Flight Distance'] = 1
data.loc[(data['Flight Distance'] > 2507.0) & (data['Flight Distance'] <= 3745.0), 'Flight Distance'] = 2
data.loc[data['Flight Distance'] > 3745.0, 'Flight Distance']                                          = 3

# Mapping Departure Delay in Minutes
data.loc[data["Departure Delay in Minutes"] <= 265.333, "Departure Delay in Minutes"] = 0
data.loc[(data["Departure Delay in Minutes"] > 265.333) & (data["Departure Delay in Minutes"] <= 530.677), "Departure Delay in Minutes"] = 1
data.loc[(data["Departure Delay in Minutes"] > 530.677) & (data["Departure Delay in Minutes"] <= 796.0), "Departure Delay in Minutes"] = 2
data.loc[(data["Departure Delay in Minutes"] > 796.0) & (data["Departure Delay in Minutes"] <= 1061.333), "Departure Delay in Minutes"] = 3
data.loc[(data["Departure Delay in Minutes"] > 1061.333) & (data["Departure Delay in Minutes"] <= 1326.667), "Departure Delay in Minutes"] = 4
data.loc[data["Departure Delay in Minutes"] > 1326.667, "Departure Delay in Minutes"] = 5

# Mapping Arrival Delay in Minutes
avg = data["Arrival Delay in Minutes"].mean()
std = data["Arrival Delay in Minutes"].std()
null_count = data["Arrival Delay in Minutes"].isnull().sum()

age_null_random_list = np.random.randint(avg - std, avg + std, size=null_count)
data["Arrival Delay in Minutes"][np.isnan(data["Arrival Delay in Minutes"])] = age_null_random_list
data["Arrival Delay in Minutes"] = data["Arrival Delay in Minutes"].astype(int)

data.loc[data["Arrival Delay in Minutes"] <= 244.833, "Arrival Delay in Minutes"] = 0
data.loc[(data["Arrival Delay in Minutes"] > 244.833) & (data["Arrival Delay in Minutes"] <= 512.677), "Arrival Delay in Minutes"] = 1
data.loc[(data["Arrival Delay in Minutes"] > 512.677) & (data["Arrival Delay in Minutes"] <= 780.5), "Arrival Delay in Minutes"] = 2
data.loc[(data["Arrival Delay in Minutes"] > 780.5) & (data["Arrival Delay in Minutes"] <= 1048.333), "Arrival Delay in Minutes"] = 3
data.loc[(data["Arrival Delay in Minutes"] > 1048.333) & (data["Arrival Delay in Minutes"] <= 1316.167), "Arrival Delay in Minutes"] = 4
data.loc[data["Arrival Delay in Minutes"] > 1316.167, "Arrival Delay in Minutes"] = 5

# Mapping Type of satisfacation
data["satisfaction"] = data["satisfaction"].map( {'satisfied': 1, 'neutral or dissatisfied': 0} ).astype(int)

X = data[list(set(data.columns) - {'Unnamed: 0', 'satisfaction'})]
y = data['satisfaction']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Standardize features by removing the mean and scaling to unit variance
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Create an MLPClassifier model
mlp = MLPClassifier(hidden_layer_sizes=(64, 12),
                    max_iter=1000, random_state=42)

# Train the model on the training data
mlp.fit(X_train, y_train)
 
# Make predictions on the test data
y_pred = mlp.predict(X_test)
 
# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
